## 통합년도 물건 개수 기준 RFM

````python
# 고객번호 데이터프레임
cust_pur_b = pd.DataFrame({'고객번호':cust_pur['고객번호'].unique()})
````

````python
# RFM
r_pur = cust_pur.groupby('고객번호')[['구매분기']].max().values
f_pur = cust_pur.groupby('고객번호')[['소분류명']].count().values
m_pur = cust_pur.groupby('고객번호')[['구매금액']].sum().values
````

````python
cust_pur_b.insert(1, '최종구매분기',r_pur)
cust_pur_b.insert(2, '구매개수',f_pur)
cust_pur_b.insert(3, '구매금액',m_pur)
````

````python
cust_pur_b
````

````python
# RFM 기준점정하기
cust_pur_b['R'] = np.where(cust_pur_b['최종구매분기'] == 8, 1, 0)
cust_pur_b['F'] = np.where(cust_pur_b['구매개수']>=np.median(f_pur),1,0)
cust_pur_b['M'] = np.where(cust_pur_b['구매금액']>=np.median(m_pur),1,0)
````

````python
cust_pur_b['RFM'] = [(i,j,k) for i,j,k in zip(cust_pur_b['R'],cust_pur_b['F'],cust_pur_b['M'])]
cust_pur_b.drop('RFM',axis=1, inplace=True)
````

## 각 고객의 총 구매한 전체상품 수

````python
def user_total_products(prd):
  고객 = prd.groupby(['고객번호']).agg(구매상품수=('소분류코드','count')).reset_index()
  return 고객
````

````python
고객_train = user_total_products(prd_train)
고객_test = user_total_products(prd_test)
````

## 각 고객의 첫 구매과 마지막 구매사이의 기간

````python
def user_period(prd,고객):
  #고객 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2차프로젝트/data/고객.csv')
  고객.sort_values('고객번호')
  고객['첫구매'] = prd.groupby('고객번호').agg(첫구매 = ('구매일자','min')).values
  고객['마지막구매'] = prd.groupby('고객번호').agg(마지막구매 = ('구매일자','max')).values

  고객['첫구매'] = pd.to_datetime(고객['첫구매'])
  고객['마지막구매'] = pd.to_datetime(고객['마지막구매'])

  고객_기간 = 고객[['첫구매','마지막구매']]
  고객_기간_diff = 고객_기간.diff(axis=1)
  고객['첫-마지막'] = 고객_기간_diff['마지막구매']

  고객['첫-마지막'] = 고객['첫-마지막'].astype(str)
  고객['첫-마지막'] = 고객['첫-마지막'].str.split(' ').str[0]
  고객['첫-마지막'] = 고객['첫-마지막'].astype(int)

  고객.drop(['첫구매','마지막구매'], axis=1, inplace=True)

  return 고객
````

````python
고객_train = user_period(prd_train,고객_train)
고객_test = user_period(prd_test,고객_test)
````

## 각 고객의 연속된 두 구매사이의 기간의 평균


````python
def user_mean_days_since_prior(prd,고객):
  #고객 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2차프로젝트/data/고객.csv')
  prd.sort_values('구매일자',inplace=True)

  고객_구매일자 = pd.DataFrame(prd.groupby(['고객번호'])['구매일자'].unique())
  고객_구매일자.reset_index(inplace=True)

  index = []
  data = []
  for i,j in enumerate(고객_구매일자['구매일자']):
    index.append(i)
    data.append(j)

  sep = pd.DataFrame(data = data,index = index)
  sep.replace(np.nan,'2000-01-01',inplace=True)
  
  period = sep.diff(axis=1)
  period = period.astype(str)
  for i in range(len(period.columns)):
    period[i] = period[i].str.split(' ').str[0]

  period.replace('NaT',0,inplace=True)
  period = period.astype(int)
  for i in range(len(period.columns)):
    period[i] = np.where(period[i]<=0,np.nan,period[i])

  period_mean = period.mean(axis=1, skipna=True)

  고객['구매간격'] = period_mean

  return 고객
````

````python
고객_train = user_mean_days_since_prior(prd_train,고객_train)
고객_test = user_mean_days_since_prior(prd_test,고객_test)
````

## 각 고객의 총 구매건수

````python
def user_orders(prd,고객):
  #고객 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2차프로젝트/data/고객.csv')

  영수증고유 = pd.DataFrame(prd[['고객번호','영수증번호']].value_counts().reset_index())
  영수증개수 = 영수증고유.groupby('고객번호')[['영수증번호']].count().reset_index()

  영수증개수.sort_values('고객번호',inplace=True)
  고객['구매건수'] = 영수증개수['영수증번호']

  return 고객
````

````python
고객_train = user_orders(prd_train,고객_train)
고객_test = user_orders(prd_test,고객_test)
````

## 각 고객의 전체 주문횟수 중 특정 상품을 포함하는 비율

````python
def up_order_rate(prd,고객):
  특정상품포함 = prd.groupby(['고객번호','소분류코드'])[['영수증번호']].count().reset_index()
  총주문횟수 = 고객[['고객번호','구매건수']]

  총_특정비율 = pd.merge(특정상품포함,총주문횟수,on = '고객번호', how='outer')
  총_특정비율['특정상품포함비율'] = 총_특정비율['영수증번호'] / 총_특정비율['구매건수']

  return 총_특정비율
````

````python
def user_total_products(prd):
  고객 = prd.groupby(['고객번호']).agg(구매상품수=('소분류코드','count')).reset_index()
  return 고객
````

````python
def user_period(prd,고객):
  #고객 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2차프로젝트/data/고객.csv')
  고객.sort_values('고객번호')
  고객['첫구매'] = prd.groupby('고객번호').agg(첫구매 = ('구매일자','min')).values
  고객['마지막구매'] = prd.groupby('고객번호').agg(마지막구매 = ('구매일자','max')).values

  고객['첫구매'] = pd.to_datetime(고객['첫구매'])
  고객['마지막구매'] = pd.to_datetime(고객['마지막구매'])

  고객_기간 = 고객[['첫구매','마지막구매']]
  고객_기간_diff = 고객_기간.diff(axis=1)
  고객['첫-마지막'] = 고객_기간_diff['마지막구매']

  고객['첫-마지막'] = 고객['첫-마지막'].astype(str)
  고객['첫-마지막'] = 고객['첫-마지막'].str.split(' ').str[0]
  고객['첫-마지막'] = 고객['첫-마지막'].astype(int)

  고객.drop(['첫구매','마지막구매'], axis=1, inplace=True)

  return 고객
````

````python
고객 = user_period(prd,고객)
````

````python
def user_mean_days_since_prior(prd,고객):
  #고객 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2차프로젝트/data/고객.csv')
  prd.sort_values('구매일자',inplace=True)

  고객_구매일자 = pd.DataFrame(prd.groupby(['고객번호'])['구매일자'].unique())
  고객_구매일자.reset_index(inplace=True)

  index = []
  data = []
  for i,j in enumerate(고객_구매일자['구매일자']):
    index.append(i)
    data.append(j)

  sep = pd.DataFrame(data = data,index = index)
  sep.replace(np.nan,'2000-01-01',inplace=True)
  
  period = sep.diff(axis=1)
  period = period.astype(str)
  for i in range(len(period.columns)):
    period[i] = period[i].str.split(' ').str[0]

  period.replace('NaT',0,inplace=True)
  period = period.astype(int)
  for i in range(len(period.columns)):
    period[i] = np.where(period[i]<=0,np.nan,period[i])

  period_mean = period.mean(axis=1, skipna=True)

  고객['구매간격'] = period_mean

  return 고객
````

````python
고객 = user_mean_days_since_prior(prd,고객)
````

````python
def user_orders(prd,고객):
  #고객 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2차프로젝트/data/고객.csv')

  영수증고유 = pd.DataFrame(prd[['고객번호','영수증번호']].value_counts().reset_index())
  영수증개수 = 영수증고유.groupby('고객번호')[['영수증번호']].count().reset_index()

  영수증개수.sort_values('고객번호',inplace=True)
  고객['구매건수'] = 영수증개수['영수증번호']

  return 고객
````
